### 9.2 Define a generator function `get_odds()`
Return the odd numbers from `range(10)` and use a loop to print the third value returned.


```python
def get_odds():
    for num in range(10):
        if num % 2 == 1:
            yield num

count = 0
for odd in get_odds():
    count += 1
    if count == 3:
        print(odd)
        break
```
