### Sorting an Array of 0s, 1s, and 2s (Dutch National Flag Algorithm)

We need to sort an array that contains only 0s, 1s, and 2s **without using the built‑in sort function**.

This solution uses the **Dutch National Flag algorithm**, which:
- Runs in **one pass** (O(n))
- Uses **constant extra space** (O(1))
- Rearranges elements by maintaining three pointers: `low`, `mid`, and `high`.

The final array will have all 0s first, then 1s, then 2s.


```python
def sort012(arr):
    low, mid, high = 0, 0, len(arr) - 1

    while mid <= high:
        if arr[mid] == 0:
            arr[low], arr[mid] = arr[mid], arr[low]
            low += 1
            mid += 1

        elif arr[mid] == 1:
            mid += 1

        else:  # arr[mid] == 2
            arr[mid], arr[high] = arr[high], arr[mid]
            high -= 1

    return arr

# Example usage:
arr = [0, 1, 2, 0, 1, 2]
sort012(arr)
```

### Binary Search — First Occurrence of k in a Sorted Array

We are given a sorted array `arr[]` and an integer `k`.  
Our task is to return the **0-based index** where `k` first appears.  
If `k` is not present, we return `-1`.

To correctly handle duplicates, we modify binary search:
- When we find `k`, we store the index,
- Then continue searching **left** to ensure we return the smallest index.

**Time Complexity:** O(log n)  
**Space Complexity:** O(1)


```python
def binary_search_first(arr, k):
    low, high = 0, len(arr) - 1
    result = -1

    while low <= high:
        mid = (low + high) // 2

        if arr[mid] == k:
            result = mid        # possible answer
            high = mid - 1      # keep searching left

        elif arr[mid] < k:
            low = mid + 1

        else:
            high = mid - 1

    return result


# Example tests
print(binary_search_first([1, 2, 3, 4, 5], 4))          # 3
print(binary_search_first([11, 22, 33, 44, 55], 445))   # -1
print(binary_search_first([1, 1, 1, 1, 2], 1))          # 0
```
