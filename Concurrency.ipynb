### 13.1 Write the current date as a string to the text file `today.txt.`

In this step, we import the `datetime` module, obtain today's date, convert it to a string, and write it to a text file named `today.txt`.


```python
from datetime import date

today = date.today()

with open("today.txt", "w") as file:
    file.write(str(today))

print(today)

```

### 13.2 Read the text file `today.txt` into the string `today_string`

This step opens the file created previously, reads its contents, and stores the date string in the variable `today_string`.


```python
# Read the contents of today.txt into today_string
with open("today.txt", "r") as file:
    today_string = file.read()

today_string
```

### 13.3 Parse the date from `today_string`

In this step, we convert the string we read from `today.txt` back into a `date` object.  
Since the string is in the ISO format `YYYY-MM-DD`, we can use `datetime.strptime` to parse it.


```python
from datetime import datetime

# Parse the date string into a date object
parsed_date = datetime.strptime(today_string, "%Y-%m-%d").date()

parsed_date
```
