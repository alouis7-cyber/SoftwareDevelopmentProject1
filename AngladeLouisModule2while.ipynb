```python
# Name: Anglade Louis
# File Name: student_honors.py
# Description: This program accepts student names and GPAs, then determines
#              whether each student qualifies for the Dean's List or Honor Roll.

def main():
    while True:
        # Ask for last name
        last_name = input("Enter student's last name (or 'ZZZ' to quit): ")
        
        # Check for sentinel value to quit
        if last_name.upper() == 'ZZZ':
            print("Exiting program.")
            break
        
        # Ask for first name
        first_name = input("Enter student's first name: ")
        
        # Ask for GPA
        try:
            gpa = float(input("Enter student's GPA: "))
        except ValueError:
            print("Invalid GPA. Please enter a numeric value.\n")
            continue
        
        # Check qualifications
        if gpa >= 3.5:
            print(f"{first_name} {last_name} has made the Dean's List.")
        elif gpa >= 3.25:
            print(f"{first_name} {last_name} has made the Honor Roll.")
        else:
            print(f"{first_name} {last_name} did not qualify for honors.")
        
        print()  # Blank line for readability


# Run the program
if __name__ == "__main__":
    main()
```
