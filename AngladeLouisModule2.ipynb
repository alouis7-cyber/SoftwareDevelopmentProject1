## Classifying an Object by Size and Color
## Exercice 4.2

We assign boolean values (`True` or `False`) to the variables **small** and **green**.  
Then we use conditional statements to determine which object matches those traits:

- **cherry** → small + red  
- **pea** → small + green  
- **watermelon** → large + green  
- **pumpkin** → large + orange  

Run the code cell below and change the values of `small` and `green` to test different outcomes.


```python
small = True
green = True

if small and green:
    print("pea")
elif small and not green:
    print("cherry")
elif not small and green:
    print("watermelon")
else:
    print("pumpkin")
```

### While loop comparison example
## Exercice 6.2

This loop compares `number` with `guess_me` and prints:
- "too low" if `number` is less than `guess_me.`
- "found it!" If they are equal, then it stops
- "oops" if `number` is greater, then stops

`number` increments at the end of each loop iteration.


```python
guess_me = 7
number = 1

while True:
    if number < guess_me:
        print('too low')
    elif number == guess_me:
        print('found it!')
        break
    else:
        print('oops')
        break

    number += 1
```

### For‑loop comparison example
## Exercice 6.3

This loop compares each `number` in `range(10)` with `guess_me = 5` and prints:
- "too low" if `number` is less than `guess_me.`
- "found it!" If they are equal, then it stops the loop
- "oops" if `number` is greater, then stops the loop


```python
guess_me = 5

for number in range(10):
    if number < guess_me:
        print('too low')
    elif number == guess_me:
        print('found it!')
        break
    else:
        print('oops')
        break
```
